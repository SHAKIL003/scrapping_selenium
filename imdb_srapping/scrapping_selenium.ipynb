```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import pandas as pd
import openpyxl

website = 'https://www.imdb.com/chart/top/?sort=rank%2Casc'
# path = r'\chromedriver.exe'
service = Service(executable_path="chromedriver.exe")
driver = webdriver.Chrome(service=service)
driver.get(website)

# to get movies by ranking
# '//option[@value="Popularity"]'

# Below code is used to click on a button
# ranking =  driver.find_element(By.XPATH,'//option[@value="Popularity"]')
# ranking.click()

excel = openpyxl.Workbook()
print(excel.sheetnames)
sheet = excel.active
sheet.title = "Top 25 Movies"
print(excel.sheetnames)
sheet.append(['Movies Names'])

# sheet.append(['Movies Rank', 'Movie Name', 'Release Year', 'Movie Rating'])


data = []
# names = driver.find_elements(By.TAG_NAME,'h3')
# years = driver.find_elements(By.TAG_NAME,'span')
names = driver.find_elements(By.TAG_NAME,'li')
for name in names:
    sheet.append([name.text])
    print(name.text)

excel.save('IMDB Movies Rating.xlsx')

# df = pd.DataFrame(data)
# driver.quit()
```
